```python
import seaborn as sns
import matplotlib.pyplot as plt
  

data = sns.load_dataset("mpg")
  

sns.regplot(x = "mpg", 
            y = "acceleration", 
            data = data)
  

plt.show()
```

```python
data = sns.load_dataset("exercise")
  

sns.regplot(x = "id",
            y = "pulse", 
            data = data)
  

plt.show()
```
